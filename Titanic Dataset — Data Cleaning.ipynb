```python
#Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Titanic dataset
df = sns.load_dataset('titanic')

# Display first few rows
df.head()
```

```python
# Overview of dataset
df.info()

# Summary statistics
df.describe(include='all')
```

```python
# Count missing values
df.isnull().sum()

# Fill 'age' with median
df['age'].fillna(df['age'].median(), inplace=True)

# Fill 'embarked' with mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

# Drop 'deck' due to too many missing values
df.drop(columns=['deck'], inplace=True)

# Drop rows where 'embark_town' or 'embarked' is still missing (if any)
df.dropna(subset=['embark_town', 'embarked'], inplace=True)
```

```python
# Drop duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Convert 'sex', 'embarked', 'class', and 'who' to category type
categorical_columns = ['sex', 'embarked', 'class', 'who']
for col in categorical_columns:
    df[col] = df[col].astype('category')
```

```python
# If 'name' column exists, extract titles
if 'name' in df.columns:
    df['Title'] = df['name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
    df['Title'] = df['Title'].replace(['Mlle', 'Ms'], 'Miss')
    df['Title'] = df['Title'].replace(['Mme', 'Lady', 'Countess'], 'Mrs')
    df['Title'] = df['Title'].replace(['Capt', 'Col', 'Major', 'Dr', 'Rev', 'Sir', 'Don', 'Jonkheer'], 'Rare')
```

```python
# Create age bins
df['age_group'] = pd.cut(df['age'], bins=[0, 12, 18, 35, 60, 100], labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
```

```python
# One-hot encoding for 'sex', 'embarked', 'class', 'age_group'
df = pd.get_dummies(df, columns=['sex', 'embarked', 'class', 'age_group'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['fare_scaled'] = scaler.fit_transform(df[['fare']])
```
